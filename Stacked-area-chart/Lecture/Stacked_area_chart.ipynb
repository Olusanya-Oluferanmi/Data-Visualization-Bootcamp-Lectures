```python
#importation of necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#Importation of csv file
df_fuel_engine_types = pd.read_csv('stacked_area_chart_data.csv')
```

```python
#To increase plot size
plt.figure(figsize=(12, 6))
#To plot x and y axis
plt.stackplot(df_fuel_engine_types['Year'],
              df_fuel_engine_types['Diesel'],
              df_fuel_engine_types['Petrol'],
              df_fuel_engine_types['Gas'])
#is used to get and set the current tick locations and labels of the x-axis
plt.xticks(df_fuel_engine_types['Year'], rotation=55)
plt.show()
```
